```python
# method overloading and super keyword
class Phone:
    def __init__(self,price,brand,camera):
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("you can buy a phone")
class Smartphone(Phone):
    def buy(self):
        print("I want to buy Iphone")
        # using the super keyword you can use the parent method in your child class
        super().buy()
    

s = Smartphone(20000, 'oppo','sony')
s.buy()
```

```python
# multilevel inheritance
class Product:
    def review(self):
        print("This is product review")

class Phone(Product):
    def __init__(self,price,brand,camera):
        self.price = price
        self.brand = brand
        self.camera= camera
    def buy(self):
        print("You can buy a phone")

class Smartphone(Phone):
    pass


s = Smartphone(2000,'iphone','sony')
s.buy()
s.review()
```

```python
# Hierarchical Inheritance
class Phone:
    def __init__(self,price,brand,camera):
        print("inside the phone class constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("you buy a phone")

class SmartPhone(Phone):
    pass

class FeaturePhone(Phone):
    pass

SmartPhone(20000,'iphone','sony').buy()
FeaturePhone(20000,'oppo','sony')
```

```python
# muliple inheritance
class Product:
    def review(self):
        print("This is product review")

class Phone:
    def __init__(self,price,brand,camera):
        self.price = price
        self.brand = brand
        self.camera= camera
    def buy(self):
        print("You can buy a phone")

class Smartphone(Phone,Product):
    pass


s = Smartphone(2000,'iphone','sony')
s.buy()
s.review()
```
